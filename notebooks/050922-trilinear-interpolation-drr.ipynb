```julia
using Plots

using DRR
```

```julia
volume, ΔX, ΔY, ΔZ = read_dicom("../data/cxr")
grid, pixels = volume2grid(volume, ΔX, ΔY, ΔZ)

# Define the camera
center = Vec3(-300., -300., -300)
camera = Camera(center)

# Define the detector plane
center = Vec3(400., 400., 400.)
normal = Vec3(-1., -1., -1.)
height, width = 201, 201
Δx, Δy = 1.5, 1.5
detector = Detector(center, normal, height, width, Δx, Δy)

# Generate the DRR
spacing = 0.1
drr = make_drr(grid, pixels, camera, detector, spacing)
heatmap(drr, c=:grays)
```

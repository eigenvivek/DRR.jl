```julia
using Plots

using DRR
```

```julia
volume, X, Y, Z = read_dicom("../data/cxr")
grid, pixels = volume2grid(volume, X, Y, Z)

# Define the camera
center = Vec3(-300., -300., -300)
camera = Camera(center)

# Define the detector plane
center = Vec3(400., 400., 400.)
normal = Vec3(-1., -1., -1.)
height, width = 201, 201
x, y = 1.5, 1.5
detector = Detector(center, normal, height, width, x, y)

# Generate the DRR
spacing = 0.1
drr = make_drr(grid, pixels, camera, detector, spacing)
heatmap(drr, c=:grays)
```
